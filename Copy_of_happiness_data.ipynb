Arranging all the rows in Ascending order on by considering the value of economy(GDP per capita) and health(Life Expectancy). 

```python
#sorting rows by Economy and Health in ascending order
import pandas as pd
def get_region():
  url=  'https://github.com/futureCodersSE/working-with-data/blob/main/Happiness-Data/2015.xlsx?raw=true'
  happiness = pd.read_excel(url)
  sort=happiness.sort_values(by =["Economy (GDP per Capita)","Health (Life Expectancy)"])
  
  return sort
get_region()
```

Identifying coorelation between different columns in dataframe 
1. Happiness score with economy(GDP per cpita)


```python
import pandas as pd 
import numpy as np
def correlation():
  url='https://github.com/futureCodersSE/working-with-data/blob/main/Happiness-Data/2015.xlsx?raw=true'
  df=pd.read_excel(url)
  col1, col2 = "Happiness Score", "Economy (GDP per Capita)"
  corr = df[col1].corr(df[col2])
  print ("Correlation between ", col1, " and ", col2, "is: ", round(corr, 2))
correlation()
```
