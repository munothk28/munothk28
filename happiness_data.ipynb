```python

#highest happiness score country

import pandas as pd
def get_data():
  url='https://github.com/munothk28/munothk28/blob/main/Happiness_dataset.xlsx?raw=true'
  happiness = pd.read_excel(url)
  sorted_rank1=happiness.sort_values("Happiness Score", ascending = False).head(1)
  display(sorted_rank1)

  
get_data()
```

```python
#sorting rows by Economy and Health in ascending order
import pandas as pd
def get_region():
  url=  'https://github.com/futureCodersSE/working-with-data/blob/main/Happiness-Data/2015.xlsx?raw=true'
  happiness = pd.read_excel(url)
  sort=happiness.sort_values(by =["Economy (GDP per Capita)","Health (Life Expectancy)"])
  
  return sort
get_region()
```
